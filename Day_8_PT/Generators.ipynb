## A generator function is a function which returns the generator- iterator with the help of yield function.

```python
x = (1, 2, 3, 4, 5)
x_com = (i*i for i in x)
print(x_com)
```

```python
def sq_num(my_list):
  res = []
  for i in my_list:
    res.append(i*i)
  return res

my_list = [1, 2, 3, 4]
print(sq_num(my_list))
```

```python
def sq_num(my_list):
  res = []
  for i in my_list:
    yield(i*i)

my_list = sq_num([1, 2, 3, 4])
print(next(my_list))
print(next(my_list))
print(next(my_list))
```

## Fibonacci Series Using Generators

```python
def fib_series(a):
  x, y = 1, 0
  while True:
    z = x + y
    if z < a:
      yield z
      x, y = y, z
    else:
      break

gen_fib = fib_series(10)
print(next(gen_fib))
print(next(gen_fib))
print(next(gen_fib))
print(next(gen_fib))
```
