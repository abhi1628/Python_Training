## Try Except Block

```python
a = 5
b = 2
try:
  print(a/b)
except Exception:
  print('Divison By 0 is not allowed')

print('Well Done')
```

```python
a = 5
b = 0
try:
  print(a/b)
except Exception as e:
  print('The error is:',e)

print('Well Done')
```

```python
try:
  a = int(input('Enter some word:'))
  print(a)
except Exception as e:
  print('The error is:',e)

print('Well Done')
```

```python
a, b = 10, 0
try:
  x = int(input('Enter some number:'))
  print(x)
  print(a/b)
except ZeroDivisionError as e:
  print('The error is:',e)
except ValueError as e:
  print('Input is not valid')
except Exception as e:
  print('The error is:',e)
```

## Try Except Finally Block

```python
a, b = 10, 0
try:
  x = int(input('Enter some number:'))
  print(x)
  print(a/b)
except ZeroDivisionError as e:
  print('The error is:',e)
except ValueError as e:
  print('Input is not valid')
except Exception as e:
  print('The error is:',e)

finally:
  print('Exception handled')
```

```python
while True:
  try:
    x = int(input('Enter some number:'))
    break
  except ZeroDivisionError as e:
    print('The error is:',e)
  except ValueError as e:
    print('Input is not valid')
  except Exception as e:
    print('The error is:',e)

  finally:
    print('Exception handled')
```

```python
x = 0
while x < 5:
  try:
    if (x != 0):
      print('x = {0}'.format(x))
  except:
      continue
  finally:
      x += 1
```

```python
def func():
  for num in range(5):
    try:
      print(num)
    finally:
      continue
func()
```
